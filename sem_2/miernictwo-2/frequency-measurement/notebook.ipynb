```python
import numpy as np
import matplotlib.pyplot as plt

def degrees_to_radians(degrees):
    radians = np.radians(degrees)
    return radians

def plot_sine_with_phase_shift(amplitude, shift):
    frequency = 1000.0  # Frequency of 1 kHz
    period = 1 / frequency
    # Generate x values for one period with a step size of 0.0001
    x = np.arange(0, period * 2, 0.00001)
    
    # Calculate y values for the normal sine function
    y_normal = amplitude * np.sin(2 * np.pi * frequency * x)
    
    # Calculate y values for the shifted sine function
    phase_shift = degrees_to_radians(shift)
    y_shifted = amplitude * np.sin(2 * np.pi * frequency * x + phase_shift)
    
    # Plot the normal sine function in blue
    plt.plot(x, y_normal, color='blue', label='sygnał wejściowy')
    
    # Plot the shifted sine function in red
    plt.plot(x, y_shifted, color='red', label=f'sygnał wyjściowy z przesunięciem fazowym {shift}°')
    
    # Set the x-axis label
    plt.xlabel('t (ms)')
    
    # Set the y-axis label
    plt.ylabel('U (V)')
    
    # Get the current x-axis ticks and labels
    x_ticks, x_tick_labels = plt.xticks()
    
    # Replace non-standard minus sign character with a standard hyphen minus
    x_tick_labels = [label.get_text().replace('−', '-') for label in x_tick_labels]
    
    # Format the tick labels to display values in multiples of 1000 (ms)
    x_tick_labels = ['{:0.2f}'.format(float(label) * 1000) for label in x_tick_labels]
    
    # Set the formatted tick labels
    plt.xticks(x_ticks, x_tick_labels)
    plt.legend(loc='upper right')
    
    # Display the plot
    plt.show()

# Example usage
amplitude = 5
frequency = 1_000
phase_shift = np.pi/2  # Phase shift of pi/2 radians (90 degrees)

```

```python
SHIFTS = [0, 326, 270, 220]

for shift in SHIFTS:
    plot_sine_with_phase_shift(amplitude, shift)
```
